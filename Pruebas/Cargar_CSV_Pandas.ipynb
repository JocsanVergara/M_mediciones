Como ordenar columnas

```python
# Ordenando de modo ascendente: de menor a mayor
df.sort_values('RSSI_1', ascending = True)
```

```python
# Ordenar de forma descendente de mayor a menor
df.sort_values('RSSI_1', ascending = False)
```

```python
# Ordenar por dos columnas
df.sort_values(['RSSI_1','Ang.Azimuth'], ascending = ['True','False'])
```
